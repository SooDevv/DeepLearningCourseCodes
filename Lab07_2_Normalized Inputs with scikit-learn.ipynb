### 학습률을 잘 조절했음에도 불구하고 NaN
- INPUT 데이터들 사이의 균형이 안맞음  
- Normalized input 으로 조정할 필요가 있다 
1. Min-max scaler 로 normalized input (0~1)

```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
X = tf.placeholder(tf.float32, shape=[None, 4])
Y = tf.placeholder(tf.float32, shape=[None, 1])
W = tf.Variable(tf.random_normal([4, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
h = tf.matmul(X, W) + b
# mse 
cost = tf.reduce_mean(tf.square(h - Y))
# minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        if step % 500 == 0:
            cost_val, h_val, _ = sess.run(
                [cost, h, train], feed_dict={X:x_data, Y: y_data})
            print('STEP : {0} , Cost: {1} \n Prediction: \n {2}'.format(step, cost_val, h_val))
```
